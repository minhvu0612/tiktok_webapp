```python
import numpy as np
import pandas as pd
import random
```

```python
gender = ["male", "female"]
```

```python
means = ["male_t", "male_f", "female_t", "female_f"]
```

```python
arr1 = []
arr2 = []
arr3 = []
for i in range(0, 400):
    arr1.append(gender[random.randrange(0,2)])
    arr2.append(random.randrange(16,50))
    arr3.append(means[random.randrange(0,4)])
```

```python
data = {
    'gender': arr1,
    'age': arr2,
    'follow': arr3,
}
```
